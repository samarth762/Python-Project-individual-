```python
# Q. print first and last five rows
import csv 
filename = "C:\\Users\\SAMARTH\\Desktop\\enol1.csv" 
fields = [] 
rows = [] 
with open(filename, 'r') as csvfile: 
    # creating a csv reader object 
    reader = csv.reader(csvfile) 
    for row in reader: 
        rows.append(row) 
n=1
print("First five rows:\n")
for row in rows[:5]:
    print(n,end=".  ")
    for col in row:
        print(col,end="  ||  ")
    n+=1
    print('\n')
print("Last five rows:\n")
for row in rows[-5:]:
    for col in row:
        print(col,end="  ||  ")
    print('\n')
```

```python
# Q. Find person who joined after 2000
import pandas as pd
df=pd.read_csv("C:\\Users\\SAMARTH\\Desktop\\enol1.csv")
a1=[]
a2=[]
n=0
joining_date=list(df.Date_of_Joining)
for i in joining_date:
    if ('/' in i):
        a1=list(map(int,i.split('/')))
        if (a1[2]>2000):
            a2.append(n)
        a1.clear()
    if ('-' in i):
        a1=list(map(int,i.split('-')))
        if (a1[2]>2000):
            a2.append(n)
        a1.clear()
    n+=1
name=list(df.First_Name)
l1=list()
for i in a2:
    l1.append(name[i])
print(list(set(l1)),"\n")
print("Total number of person joined after '2000' :",len(list(set(l1))))

```

```python
salary=list(df.Salary)
name=list(df.First_Name)
l2=list()
for i in salary:
    if (int(i)>100000):
        l2.append(salary.index(i))
n=1
l3=list()
for i1 in l2:
    l3.append(name[i1])
for i2 in list(set(l3)):
    print(n,". ",i2)
    n+=1
```
